# **Feature Scaling**

Feature scaling is a technique used to normalize the range of independent variables in a dataset. It is a process of scaling the features of the data to a common range. 
This is done to prevent features with large ranges from dominating the model. There are several methods for feature scaling, including:
- **Standardization**: This method subtracts the mean and divides by the standard deviation for each feature . This is the most common method of feature scaling. 
- **Normalization**: This method scales the features to a common range, usually between 0 and 1. This is useful when the features have different units or scales.
- **Min-Max Scaling**: This method scales the features to a common range, usually between 0 and 1. This is similar to normalization, but it uses the minimum and maximum values of each feature to scale the data.

Feature scaling is an important step in machine learning because it can improve the performance of the model. It can also help to prevent features with large ranges from dominating the model. 
This is particularly useful when using distance-based algorithms, such as k-Nearest Neighbors (k-NN). The goal of feature scaling is to ensure that all features are on the same scale, so that no single feature dominates the others.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://github.com/RyanNolanData/YouTubeData/blob/main/500hits.csv?raw=true", encoding="latin-1")
df.head()
```

```python
df = df.drop(columns = ["PLAYER", "CS"])
```

```python
df. describe().round(2)
```

```python
X1 = df.iloc[:, 0:13]
X2 = df.iloc[:, 0:13]
```

For Normalization in Feature Scaling, scikit-learn uses the below formula:-
* (X - Min(X))/(Max(X) - Min(X))
- It is bounded between 0 and 1.

For Standardization in Feature Scaling, scikit-learn uses the below formula:-
* (X - Mean(X))/Standard Deviation(X)
- Unlike Normalization, Standardization is not bounded between 0 and 1 and can have +ve and -ve numbers larger than 1.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaleStandard = StandardScaler()
X1 = scaleStandard.fit_transform(X1) 
# fit_transform() is used to fit the data and then transform it(standardize it).
# If we use only transform() it will not fit the data and will not standardize it
X1 = pd.DataFrame(X1, columns= df.columns[0:13])
```

```python
X1.head()
```

```python
X1.describe().round(3)
# This shows that the mean is 0 and standard deviation is 1, which is the property of Standardization.
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaleMinMax = MinMaxScaler(feature_range=(0, 1))
X2 = scaleMinMax.fit_transform(X2)
# fit_transform() is used to fit the data and then transform it(normalize it).
X2 = pd.DataFrame(X2, columns=df.columns[0:13])
```

```python
X2.head()
```

```python
X2.describe().round(3)
# This shows that the minimum value is 0 and maximum value is 1, which is the property of Min-Max Scaling.
```
